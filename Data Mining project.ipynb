```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
salarydata = pd.read_csv('sample_ds_salary.csv')
```

```python
salarydata
```

```python
#Bar plot for company_size counts
plt.figure(figsize=(7, 5))
sns.countplot(data=salarydata, x='company_size')
plt.title('Distribution of the Company Size')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

```python
#Histoplot showing Salary distribution
plt.figure(figsize=(7, 5))
sns.histplot(data=salarydata, x='salary', kde=True)
plt.title('Distribution based on salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

```python
#Box plot showing salary vs employment_type
plt.figure(figsize=(12, 8))
sns.boxplot(data=salarydata, x='employment_type', y='salary')
plt.title('Box plot showing salary vs employment_type')
plt.xlabel('Employment Type')
plt.ylabel('Salary')
plt.show()
```

```python
#Scatter plot for salary vs. work_year
plt.figure(figsize=(12, 8))
sns.scatterplot(data=salarydata, x='work_year', y='salary', hue='work_year')
plt.title('Scatterplot for Salary vs. Work Year')
plt.xlabel('Work Year')
plt.ylabel('Salary')
plt.legend(title='Work Year')
plt.show()
```
